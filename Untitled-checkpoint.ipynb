```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Generate Synthetic Dataset with Outliers
np.random.seed(42)

# Generating dates from January 1, 2022, to December 31, 2022
dates = pd.date_range(start="2015-01-01", end="2022-12-31", freq="W")

# Generating synthetic sales data with a trend, some noise, and outliers
sales = np.arange(1, len(dates) + 1) * 5 + np.random.normal(scale=20, size=len(dates))

# Introduce outliers at specific indices
outlier_indices = np.random.choice(len(dates), size=5, replace=False)
sales[outlier_indices] += 200  # Increase sales for outliers

# Creating a DataFrame
df = pd.DataFrame({"Date": dates, "Sales": sales})

# Data Visualization
plt.figure(figsize=(10, 6))
plt.plot(df["Date"], df["Sales"], label="Actual Sales")
plt.title("Synthetic Daily Sales Data with Outliers")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.show()
```
